## 1. Login

```python
driver = webdriver.Chrome('../chromedriver') #Set path to web driver
driver.maximize_window()

driver.get('https://www.instagram.com/?hl=en')
wait = WebDriverWait(driver, 10)#driver, upper limit
login = wait.until(ec.presence_of_element_located((By.LINK_TEXT,"Log in")))
login.click()
userfield = wait.until(ec.presence_of_element_located((By.XPATH,"//input[@aria-label = 'Phone number, username, or email']")))
passfield = wait.until(ec.presence_of_element_located((By.XPATH,"//input[@name = 'password']")))

userfield.send_keys(sample_username)
passfield.send_keys(sample_password)

btn = driver.find_element_by_tag_name('button')
btn.submit()
time.sleep(2)
try:
    driver.find_element_by_xpath("//*[contains(text(), 'Not Now')]").click()
except NoSuchElementException:
    pass
```

## 5. Like/Unlike posts
Liking the top 30 posts of the ‘dilsefoodie'. Print message if already liked.
Unliking the top 30 posts of the ‘dilsefoodie’. Print message if already unliked.

```python
t = driver.find_element_by_xpath("//input[@placeholder = 'Search']") # to avoid StaleElementReferenceException
t.send_keys("dilsefoodie")

element = wait.until(ec.presence_of_element_located((By.CLASS_NAME,'yCE8d ')))
driver.get(element.get_attribute('href'))
time.sleep(2)

current_height = driver.execute_script('return document.body.scrollHeight;')
while(len(driver.find_elements_by_xpath("//div[@class = 'Nnq7C weEfm']"))<10): # scroll 10 rows for 30 images
    driver.execute_script('window.scrollBy(0,arguments[0]);', current_height)
    new_height = driver.execute_script('return document.body.scrollHeight;')
    current_height = new_height

rows = driver.find_elements_by_xpath("//div[@class = 'Nnq7C weEfm']")

hrefs = [] #collecting links for top 30 posts
for row in rows:
    posts = row.find_elements_by_xpath("./*")
    for post in posts:
        link = post.find_element_by_tag_name('a')
        hrefs.append(link.get_attribute('href'))
```

```python
# Liking 30 posts
count = 0
for pic in hrefs:
    driver.get(pic)
    xp = "//*[name()='svg']"
    svg = wait.until(ec.presence_of_element_located((By.XPATH,xp)))
#     svg = driver.find_element_by_xpath(xp)
    label = svg.get_attribute('aria-label')
    if label=='Unlike':
        print("Already Liked")
    else:
        btn = svg.find_element_by_xpath("./..")
        btn.click()
        count+=1
        print("Liked post ",count) #comment for cleaner output
```

```python
# Unlike liked images
count=0
for pic in hrefs:
    driver.get(pic)
    xp = "//*[name()='svg']"
    svg = wait.until(ec.presence_of_element_located((By.XPATH,xp)))
    label = svg.get_attribute('aria-label')
    if label=='Unlike':
        btn = svg.find_element_by_xpath("./..")
        btn.click()
        count+=1
        print("Unliked post ",count) #comment for cleaner output
    else:
        print("Already Unliked")
```

## 6. Extract list of followers
Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
profiles = ['foodtalkindia','sodelhi']
followers = []
for profile in profiles:
    t = wait.until(ec.presence_of_element_located((By.XPATH,"//input[@placeholder = 'Search']"))) 
    t.send_keys(profile)
    xp = "//a[@href = '/" + profile + "/']"
    follxp = "//a[@href = '/" + profile + "/followers/']"
    element = wait.until(ec.presence_of_element_located((By.XPATH,xp)))
    element.click()
    element = wait.until(ec.presence_of_element_located((By.XPATH,follxp)))
    element.click()

    dialog = wait.until(ec.presence_of_element_located((By.XPATH,"//div[@role = 'dialog']")))
    div = wait.until(ec.presence_of_element_located((By.CLASS_NAME,'PZuss')))

    fols = wait.until(ec.presence_of_element_located((By.CLASS_NAME,'isgrP')))
    current_count = len(div.find_elements_by_tag_name("li"))
    
    while current_count < 500:
        driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", fols)
        new_count = len(div.find_elements_by_tag_name("li"))
        time.sleep(1)
        current_count = new_count
    
    div = driver.find_element_by_class_name('PZuss')
    followers_list = div.find_elements_by_xpath("./li")
    followers.append([])
    
    namepath = 'div/div[1]/div[2]/div[1]/a'
    for pf in followers_list:
        name = pf.find_element_by_xpath(namepath).text
        followers[profiles.index(profile)].append(name)
    
    close = driver.find_element_by_xpath("//*[@aria-label='Close']")
    close.click()
```
